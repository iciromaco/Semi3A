```
url = 'https://monosnap.com/image/QIIH2GJkSav4QXwJfRixhf7bF2ReAy.png' # 画像URL
import urllib.request
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img_path, header = urllib.request.urlretrieve(url)
print(header) # アクセスが拒否されていないかどうかのチェック用
img = cv2.imread(img_path,1)
cv2_imshow(img)
cv2.imwrite('lena.jpg',img)
```

```
import cv2
from matplotlib import pyplot as plt
 
fname = "lena.jpg" # ファイル名は実状に合わせる。
src_img = cv2.imread(fname,0) # グレイで読み込み
 
graph = cv2.calcHist([src_img],[0],None,[256],[0,256])
plt.plot(graph)
plt.xlim([0,256])
plt.show()
```

## カラー成分ごとのヒストグラム

```
src_img = cv2.imread(fname, -1)

colors = ['b','g','r','gray']

if len(src_img.shape) == 2: # グレイ画像だった場合
  colors = colors[:1]
  
for i in range(src_img.shape[2]):
  graph = cv2.calcHist([src_img],[i],None,[256],[0,256])
  plt.plot(graph,color=colors[i])
  
plt.xlim([0,256])
plt.show()
```
